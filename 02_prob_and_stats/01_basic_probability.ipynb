# Probability

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rolls = np.random.randint(0, 6, size=100)
counts = np.zeros((6, 100))
totals = np.zeros(6)

for i in rolls:
    totals[i] += 1
    counts[:, i] = totals

totals / 100
```

```python
x = np.arange(100).reshape(1, 100) + 1
estimates = counts / x
```

```python
plt.figure(figsize=(12, 8))

for i in range(6):
    plt.plot(estimates[i, :], label='P(die={})'.format(i + 1))

plt.axhline(y=0.1666, color='black', linestyle='dashed')
plt.legend()
plt.show()
```
